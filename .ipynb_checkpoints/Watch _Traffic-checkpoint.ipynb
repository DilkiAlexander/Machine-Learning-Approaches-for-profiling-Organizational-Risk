```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import joblib
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import mysql.connector
import subprocess

import joblib
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from tkinter import messagebox

import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.utils import formatdate

# MySQL Configuration
mysql_db_config = {
    "host": "network-anomaly-db.cromkkyk46eq.ap-south-1.rds.amazonaws.com",
    "user": "admin",
    "password": "Abc$$1234",
    "database": "network_data"
}

def read_data_in_db():
    connection = mysql.connector.connect(**mysql_db_config)
    cursor = connection.cursor()

    read_query = "SELECT info, ipaddress FROM net_queries_syn ORDER BY id DESC LIMIT 1"
    cursor.execute(read_query)

    # Fetch the result
    row = cursor.fetchone()
    if row is not None:
        log_info = row[0]
        ip_address = row[1]
    else:
        log_info = None
        ip_address = None

    cursor.close()
    connection.close()

    return log_info, ip_address

def update_status_in_db():
    connection = mysql.connector.connect(**mysql_db_config)
    cursor = connection.cursor()

    update_query = "UPDATE net_queries_syn SET status = %s WHERE id = %s"
    cursor.execute(update_query)

    connection.commit()

    cursor.close()
    connection.close()

def determine_attack_type(info):
    # Existing attack type detections
    if 'SYN' in info and 'ACK' not in info:
        return "SYN Flood (DoS Attack)"
    if 'ACK' in info and 'PSH' not in info and 'SYN' not in info:
        return "ACK Flood"
    return "Unknown"

def send_email(subject, message):
    try:
        from_email = 'idssolutions123@gmail.com'
        password = 'yyjwyodyxyyzcyii'
        to_email = 'alexande-cs18002@stu.kln.ac.lk'

        # Set up the MIME
        msg = MIMEMultipart()
        msg['From'] = from_email
        msg['To'] = to_email
        msg['Subject'] = subject
        msg['Date'] = formatdate(localtime=True)

        # Attach the message body
        msg.attach(MIMEText(message, 'plain'))

        # Establish a secure session with the server
        server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
        server.ehlo()

        # Login to the server
        server.login(from_email, password)

        # Send the email
        server.sendmail(from_email, to_email, msg.as_string())

        # Close the server connection
        server.quit()

        print("Email sent successfully.")

    except Exception as e:
        print(f"Failed to send email: {e}")

# Load the dumped models
classifier = joblib.load('decision_tree_model_optimized.joblib')
cv = joblib.load('count_vectorizer.joblib')

# Function to clean the texts
def preprocess_text(text):
    log = re.sub('[^a-zA-Z0-9]', ' ', text)
    log = log.lower()
    log = log.split()
    ps = PorterStemmer()
    log = [ps.stem(word) for word in log if not word in set(stopwords.words('english'))]
    log = ' '.join(log)
    return log

# New log insert
new_text, ip_address = read_data_in_db()
preprocessed_text = preprocess_text(new_text)

# Use the loaded CountVectorizer object to transform the preprocessed text into a bag of words representation
new_text_bow = cv.transform([preprocessed_text]).toarray()

# Use the loaded classifier to make predictions on the new log data
predicted_class = classifier.predict(new_text_bow)

# Print the log input and the predicted class
print(f"Log input: {new_text}")
print(f"Predicted class: {predicted_class}")

# Display a pop-up message based on the predicted class
if predicted_class:
    attack_type = determine_attack_type(new_text)
    email_subject = "Network Anomaly Detection"
    email_message = f"""
    An attack has been detected from the IP address: {ip_address}.

    Log Information:
    {new_text}

    Type of Attack: {attack_type}

    Risk Profile: High
    """
    send_email(email_subject, email_message)
```
