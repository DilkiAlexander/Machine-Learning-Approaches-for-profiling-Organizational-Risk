```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the provided table
models = ['Decision Tree', 'Naive Bayes', 'Random Forest', 'SVM']
original_accuracies = [97.08, 72.71, 94.00, 75.00]
optimized_accuracies = [98.67, 98.55, 95.00, 98.00]

# Define the bar width
bar_width = 0.35

# Set the positions of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar plots
plt.figure(figsize=(10, 6))
bars1 = plt.bar(r1, original_accuracies, color='b', width=bar_width, edgecolor='grey', label='Original')
bars2 = plt.bar(r2, optimized_accuracies, color='g', width=bar_width, edgecolor='grey', label='Optimized')

# Add labels
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Accuracy (%)', fontweight='bold')
plt.title('Model Performance: Original vs Optimized Accuracy')
plt.xticks([r + bar_width / 2 for r in range(len(models))], models)

# Add legend
plt.legend()

# Add the accuracy values on top of the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval, f'{yval:.2f}%', ha='center', va='bottom')  # va: vertical alignment

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval, f'{yval:.2f}%', ha='center', va='bottom')  # va: vertical alignment

# Save the plot
plt.savefig('model_performance_plot_final.png')

plt.show()
```
